#Merge dataset

```python
import os
import shutil
from tqdm import tqdm
import yaml

# --------------------------------------
# 1. Setup Paths (Colab Specific)
# --------------------------------------


# Define paths (modify these!)
indian_path = "/content/Indian-Food-Yolov5-2"
foodseg_path = "/content/foodseg103_yolo_deterministic"
output_path = "/content/merged_dataset"

# --------------------------------------
# 2. Create Merged Directory Structure
# --------------------------------------
os.makedirs(f"{output_path}/train/images", exist_ok=True)
os.makedirs(f"{output_path}/train/labels", exist_ok=True)
os.makedirs(f"{output_path}/valid/images", exist_ok=True)
os.makedirs(f"{output_path}/valid/labels", exist_ok=True)

# --------------------------------------
# 3. Merge Indian-Food Dataset (No Changes Needed)
# --------------------------------------
def copy_dataset(src_path, dest_path, dataset_type="train"):
    # Copy images
    shutil.copytree(
        f"{src_path}/{dataset_type}/images",
        f"{dest_path}/{dataset_type}/images",
        dirs_exist_ok=True
    )
    # Copy labels
    shutil.copytree(
        f"{src_path}/{dataset_type}/labels",
        f"{dest_path}/{dataset_type}/labels",
        dirs_exist_ok=True
    )

print("Copying Indian-Food dataset...")
copy_dataset(indian_path, output_path, "train")
copy_dataset(indian_path, output_path, "valid")

# --------------------------------------
# 4. Merge FoodSeg103 (With Class ID Offset)
# --------------------------------------
def process_foodseg(src_path, dest_path, offset=81):
    for dataset_type in ["train", "valid"]:
        # Images
        for img_file in tqdm(os.listdir(f"{src_path}/{dataset_type}/images")):
            shutil.copy(
                f"{src_path}/{dataset_type}/images/{img_file}",
                f"{dest_path}/{dataset_type}/images/foodseg_{img_file}"
            )

        # Labels (with class ID offset)
        for label_file in tqdm(os.listdir(f"{src_path}/{dataset_type}/labels")):
            with open(f"{src_path}/{dataset_type}/labels/{label_file}", 'r') as f:
                lines = [f"{int(line.split()[0]) + offset} {' '.join(line.split()[1:])}\n"
                        for line in f.readlines()]

            with open(f"{dest_path}/{dataset_type}/labels/foodseg_{label_file}", 'w') as f:
                f.writelines(lines)

print("\nProcessing FoodSeg103 dataset with class offset...")
process_foodseg(foodseg_path, output_path, offset=81)

# --------------------------------------
# 5. Create Merged data.yaml
# --------------------------------------
# Load original class names
with open(f"{indian_path}/data.yaml") as f:
    indian_classes = yaml.safe_load(f)['names']

with open(f"{foodseg_path}/data.yaml") as f:
    foodseg_classes = yaml.safe_load(f)['names']

merged_config = {
    'path': output_path,
    'train': 'train/images',
    'valid': 'valid/images',
    'names': indian_classes + foodseg_classes,
    'nc': len(indian_classes) + len(foodseg_classes)
}

with open(f"{output_path}/data.yaml", 'w') as f:
    yaml.dump(merged_config, f)

# --------------------------------------
# 6. Verification
# --------------------------------------
def count_files(path):
    return len([f for f in os.listdir(path) if f.endswith('.jpg') or f.endswith('.txt')])

print("\nVerification:")
print(f"Train images: {count_files(f'{output_path}/train/images')}")
print(f"Train labels: {count_files(f'{output_path}/train/labels')}")
print(f"Val images: {count_files(f'{output_path}/valid/images')}")
print(f"Val labels: {count_files(f'{output_path}/valid/labels')}")

print("\nMerging complete! Dataset structure:")
print("\n=== Directory Structure ===")
def print_dir_structure(startpath, indent=''):
    for root, dirs, files in os.walk(startpath):
        # Skip showing files if more than 5
        level = root.replace(startpath, '').count(os.sep)
        indent = ' ' * 4 * (level)
        print(f"{indent}{os.path.basename(root)}/")
        subindent = ' ' * 4 * (level + 1)
        for f in files[:5]:  # Show first 5 files only
            print(f"{subindent}{f}")
        if len(files) > 5:
            print(f"{subindent}... and {len(files)-5} more")

print_dir_structure(output_path)
```
